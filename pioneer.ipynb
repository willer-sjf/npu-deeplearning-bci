```python
def integrate_eeg_motormovement_data(PATH):
    ndata = None
    nlabel = None
    data_lens = 640
    for i in range(1, 100):
        prefix = 'S' + '{0:03d}'.format(i) 
        for j in range(3, 15):
            subfix = prefix + 'R{0:02d}'.format(j)
            file_name = PATH + prefix + '/' + subfix + '.edf'
            rdata = mne.io.read_raw_edf(file_name, preload=True)
            pdata = rdata.to_data_frame()
            events_from_annot, event_dict = mne.events_from_annotations(rdata)
            
            tdata = np.array(pdata.iloc[:data_lens, 1:]).reshape(1, -1, 64)
            for idx in range(1, len(events_from_annot)):
                sdata = np.array(pdata.iloc[events_from_annot[idx][0]:events_from_annot[idx][0]+data_lens ,1:]).reshape(1, -1, 64)
                tdata = np.concatenate([tdata, sdata], 0)
            tdata = tdata.swapaxes(1,2)
            tlabel = np.array(list(zip(*events_from_annot))[-1])
            tlabel-= 1
            if str(type(ndata)) == "<class 'NoneType'>":
                ndata  = tdata
                nlabel = tlabel
            else:
                ndata  = np.concatenate([ndata, tdata], 0)
                nlabel = np.concatenate([nlabel, tlabel], 0)
                
    np.save(PATH + 'inte_eeg-motormovement_data.npy', ndata)
    np.save(PATH + 'inte_eeg-motormovement_label.npy', nlabel)
    print("<<=== Data Saved ===>>")
```

```python
integrate_eeg_motormovement_data(PATH)
```
